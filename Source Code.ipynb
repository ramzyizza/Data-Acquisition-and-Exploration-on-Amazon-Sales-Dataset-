### Data Cleaning
We first start by checking for any duplicate rows, followed with identifying missing value on the dataset. Then replace data that has symbol and remove irrelevant and null data.

```python
#Check whether any duplicate value exist
df.duplicated().values.any()
```

```python
#Check whether any missing value exist
df.isna().values.any()
```

```python
#Check which columns that has missing value
df.isna().sum()
```

```python
#Drop column that has missing value
df.dropna(axis=0, how='any', inplace = True)
```

**We want to remove irrelevant symbols in our data, particularly attributes that we want to convert to integer by replacing currency and percentage symbol that appear in price and discount**


```python
#Remove and replace symbols for consistency matters
df.actual_price = df.actual_price.str.replace('₹' , '')
df.actual_price = df.actual_price.str.replace(',' , '')
df.discounted_price = df.discounted_price.str.replace('₹' , '')
df.discounted_price = df.discounted_price.str.replace(',' , '')
df.discount_percentage = df.discount_percentage.str.replace('%' , '')
df.rating_count = df.rating_count.str.replace(',' , '')
```

Drop non-integer value on rating that has the value that is not numeric

```python
#Check rating that is not number
df.rating.unique()
```

```python
#Locate the rows that has the rating non-numeric
df.loc[df.rating == '|']
```

```python
#Drop the row that has this value
df.drop(index=1279 , inplace= True)
```

Change numeric datatype from string to float

```python
df.discounted_price = df.discounted_price.astype('float64')
df.actual_price =  df.actual_price.astype('float64')
df.discount_percentage =df.discount_percentage.astype('float64')
df.rating_count = df.rating_count.astype('float64')
df.rating = df.rating.astype('float64')
```

## Data Visualization
This section, our objective is to describe our data in the form of visualization.

```python
# Group by category and count the number of products in each category
category_counts = df.groupby('category')['product_id'].count().sort_values(ascending=False)[:10]
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2', '#CCB974', '#64B5CD', '#4DB27D', '#E69F00', '#F0E442', '#D55E00']

# Create a pie chart of the top 10 categories
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, counterclock=False, colors=colors)
ax.set_title('Distribution of Top 10 Product Categories Based on the Number of Product Reviews', fontsize=14, fontweight='bold')
plt.show()
```

```python
# group by category and calculate mean rating
avg_rating = df.groupby('category')['rating'].mean().sort_values(ascending=False)[:50]

# create bar chart
fig, ax = plt.subplots(figsize=(20, 6))
ax.bar(avg_rating.index, avg_rating.values, color='royalblue')
ax.set_xlabel('Category', fontsize=14, fontweight='bold')
ax.set_ylabel('Average Rating', fontsize=14, fontweight='bold')
ax.set_title('Average Rating of 50 Categories', fontsize=16, fontweight='bold')
ax.set_ylim(1, 5)
ax.set_yticks([i/2 for i in range(2, 11)])
ax.set_yticklabels([str(i/2) for i in range(2, 11)])
plt.xticks(rotation=90)
plt.show()
```

```python
# Group the dataset by product name and calculate the sum of rating counts
product_ratings = df.groupby("product_name")["rating_count"].sum()
# Sort the products by rating count in descending order and select the top 10
top_products = product_ratings.sort_values(ascending=False)[:10]

# Create the bar chart
sns.barplot(x=top_products.values, y=top_products.index, color="royalblue")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")
plt.title("Top 10 Best-Selling Products with Highest Rating Counts")
plt.show()
```
